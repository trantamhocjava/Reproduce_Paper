```python
from kltn_utils import kltn_utils
from kltn_utils.path import path
import os 
```

```python
file_path = "gpt_class2concept/isic2018.json"
class2concept = kltn_utils.read_json_to_dict(file_path)

concepts = []
concept2class = []
class_names = list(class2concept.keys()) 

for class_idx, class_name in enumerate(class_names): 
    concepts_item = class2concept[class_name]
    concepts += concepts_item
    concept2class += [[class_idx]] * len(concepts_item)

print(f"class_names: {class_names}")
print(f"concepts[:10]: {concepts[:10]}")
print(f"concept2class[:10]: {concept2class[:10]}")
```

```python
file_path = "gpt_class2concept/isic2018.json"

name = os.path.splitext(os.path.basename(file_path))[0]

print(name)
```

```python
def create_json(file_path): 
    _, file_name = path.get_file_name(file_path)
    os.makedirs(f"data/{file_name}", exist_ok=True)
    
    class2concept = kltn_utils.read_json_to_dict(file_path)

    concepts = []
    concept2class = []
    class_names = list(class2concept.keys()) 

    for class_idx, class_name in enumerate(class_names): 
        concepts_item = class2concept[class_name]
        concepts += concepts_item
        concept2class += [[class_idx]] * len(concepts_item)

    result = {
        "concepts": concepts, 
        "concept2class": concept2class, 
        "class_names": class_names, 
    }

    kltn_utils.save_dict_to_json(result, f"data/{file_name}/class_concept.json")
    kltn_utils.save_dict_to_json(class2concept, f"data/{file_name}/class2concept.json")

    print("save ok")
```

```python
create_json("gpt_class2concept/isic2018.json")
```
