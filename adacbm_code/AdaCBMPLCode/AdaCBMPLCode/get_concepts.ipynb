```python
from kltn_utils import kltn_utils
from gpt_concepts.concept_dataset import isic2018_concept_dataset
```

```python
concepts = []
concept2class = []
class_names = list(isic2018_concept_dataset.keys()) 

for class_idx, class_name in enumerate(class_names)  : 
    class_concepts = isic2018_concept_dataset[class_name]
    concepts += class_concepts
    concept2class += [class_idx] * len(class_concepts)

print(f"concepts[:10]: {concepts[:10]}")
print(f"concept2class[:10]: {concept2class[:10]}")
```

```python
kltn_utils.save_dict_to_json({"concepts": concepts}, "concepts.json")
```

```python
kltn_utils.save_dict_to_json({"concept2class": concept2class}, "concept2class.json")
```

```python
kltn_utils.save_dict_to_json({"class_names": class_names}, "class_names.json")
```
